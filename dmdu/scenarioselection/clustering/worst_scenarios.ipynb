# Visualization of Worst-Case Scenarios from Time Series Clustering

```python
import pandas as pd
import seaborn as sns
import os
from ema_workbench.analysis import parcoords
import matplotlib.pyplot as plt
from dmdu.general.xlm_constants_epsilons import get_uncertainty_names
```

```python
path = os.path.join(
    os.getcwd(),
    'data',
    'time_series_scenarios.csv'
)
scenarios = pd.read_csv(path, index_col='Unnamed: 0')

# To change the order of the columns
x_names = get_uncertainty_names()
scenarios = scenarios.reindex(columns=x_names)
scenarios
```

```python
sns.set(font_scale=1.8)
sns.set_style("whitegrid")
sns.set(rc={'figure.figsize': (12, 8)})

limits = parcoords.get_limits(scenarios)
axes = parcoords.ParallelAxes(limits)
axes.plot(scenarios)
plt.show()
```

```python
path = os.path.join(
    os.path.dirname(os.path.dirname(os.getcwd())),
    'outputimages',
    'scenariodiscovery',
    'tsc_worst_scenarios_pap.png'
)
axes.fig.savefig(path, dpi=200, pad_inches=0.2, bbox_inches='tight')
```
