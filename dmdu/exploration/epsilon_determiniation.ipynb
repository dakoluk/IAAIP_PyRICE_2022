# Epsilon Determination

Here, we load some random open exploration data and analyze what epsilon values per objective seem fitting.

```python
from dmdu.general.xlm_constants_epsilons import get_all_outcome_names
from ema_workbench import load_results
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
import os
```

## Directory

```python
target_directory = os.path.join(os.getcwd(), 'data')
```

## Relevant outcomes

```python
outcome_names = get_all_outcome_names()
```

## Load Results

```python
file_name = 'open_exploration_100x100'
results = load_results(file_name=os.path.join(target_directory, file_name))

experiments, outcomes = results
outcomes = pd.DataFrame(outcomes)

names = get_all_outcome_names()
names = [n + ' 2105' for n in names]

only_2105 = outcomes.loc[:, names]
only_2105
```

# Visualization of distributions within the objectives

```python
sns.set(font_scale=1.6)
sns.set_style("whitegrid")

for name in names:
    sns.violinplot(data=only_2105.loc[:, name]).set(title=f'{name}')
    plt.show()
```

```python
only_2105.describe()
```

## Choosing epsilon values

```python
name = 'Distance to consumption threshold 2105'
table = only_2105.loc[(only_2105[name] < 0.0005) & (only_2105[name] > 0.0)]
table.loc[:, name]
```

Conclusion: $\epsilon = 0.0001$

```python
name = 'Disutility 2105'
table = only_2105.loc[(only_2105[name] < 6) & (only_2105[name] > 0.0)]
table.loc[:, name]
```

Conclusion: $\epsilon = 0.1$

```python
name = 'Utility 2105'
table = only_2105.loc[(only_2105[name] < 2512) & (only_2105[name] >= 0.0)]
table.loc[:, name]
```

Conclusion: $\epsilon = 0.05$

```python
name = 'Distance to damage threshold 2105'
table = only_2105.loc[(only_2105[name] < 0.005) & (only_2105[name] > 0.0)]
table.loc[:, name]
```

Conclusion: $\epsilon = 0.0001$

```python
name = 'Intratemporal consumption Gini 2105'
table = only_2105.loc[(only_2105[name] < 0.2786) & (only_2105[name] > 0.0)]
table.loc[:, name]
```

Conclusion: $\epsilon = 0.0001$

```python
name = 'Intratemporal damage Gini 2105'
table = only_2105.loc[(only_2105[name] < 0.5457) & (only_2105[name] > 0.0)]
table.loc[:, name]
```

Conclusion: $\epsilon = 0.0001$

```python
name = 'Lowest income per capita 2105'
table = only_2105.loc[(only_2105[name] < 2) & (only_2105[name] > 0.0)]
table.loc[:, name]
```

Conclusion: $\epsilon = 0.1$

```python
name = 'Highest damage per capita 2105'
table = only_2105.loc[(only_2105[name] < 0.0152) & (only_2105[name] > 0.0)]
table.loc[:, name]
```

Conclusion: $\epsilon = 0.0005$
