# Convergence Behavior of directed scenario search

Directed scenario search has been run for each problem formulation. The convergence behavior for `Epsilon-Porgress` and for `Hypervolume` are shown below.

```python
import os
from ema_workbench.analysis import parcoords
from dmdu.general.convergence import plot_epsilon_progress
from dmdu.scenarioselection.search.directed_scenario_search import load_scenario_search_results
import seaborn as sns
import matplotlib.pyplot as plt
```

## Indicating which data to look at

```python
data_folder_path = os.path.join(os.getcwd(), 'data')
searchover = 'uncertainties'
nfe = 200000
```

## Visualization of $\epsilon$-Progress

```python
plot_epsilon_progress(data_folder_path=data_folder_path, searchover=searchover, nfe=nfe, saving=False)
```

## Scenarios

```python
scenarios = load_scenario_search_results(data_folder_path=data_folder_path, searchover=searchover, nfe=nfe, n_references=1)
scenarios
```

```python
sns.set(rc={'figure.figsize':(20, 10)})
limits = parcoords.get_limits(scenarios)
axes = parcoords.ParallelAxes(limits)
axes.plot(scenarios)

visualization_folder = os.path.join(
    os.path.dirname(os.path.dirname(os.getcwd())),
    'outputimages'
)
plt.savefig(os.path.join(visualization_folder, 'dss_scenarios.png'), dpi=200, pad_inches=0.2)

plt.show()
```
